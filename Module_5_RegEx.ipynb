```python
txt = 'my name is sudheshna.Not Sure Where To Begin	Returns a match where the specified characters are at the beginning or at the end of a word?'
```

```python
import re
x = re.search('name',txt)
print(x.start())
```

```python
import re
x = re.findall("sudh", txt)
print(x)
```

```python
import re
x = re.search("\s", txt)

print("Not Sure Where To Begin:", x.start()) 
```

```python
import re
x = re.search("specified", txt)
print(x.start())
```

```python
#Split the string at the first white-space character:
import re
txt = "Not Sure Where To Begin"
x = re.split("\s", txt, 1)
print(x)
```

```python
#Replace all white-space characters with the digit "9":
import re
txt = "Not Sure Where To Begin"
x = re.sub("\s", "9", txt)
print(x)
```

```python
#Replace the first two occurrences of a white-space character with the digit 9:
import re
txt = "The rain in Spain"
x = re.sub("\s", "9", txt, 2)
print(x)
```

```python
#The search() function returns a Match object:
import re
txt = "Not Sure Where To Begin"
x = re.search("ai", txt)
print(x)
```

```python
#Search for an upper case "S" character in the beginning of a word, and print the word:
import re
txt = "Not Sure Where To Begin"
x = re.search(r"\bS\w+", txt)
print(x.group())
```
